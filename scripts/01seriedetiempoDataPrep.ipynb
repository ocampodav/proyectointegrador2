###**Cargar y dar formato a los datos**

##### **Órdenes de trabajo**

```python
#Cargar el registro histórico de órdenes de trabajo
#ot = pd.read_excel('/content/drive/MyDrive/ProyectoIntegrador2/exportOT.xlsx', sheet_name='Exportar Hoja de Trabajo')
ot = pd.read_excel('/content/exportOT.xlsx', sheet_name='Exportar Hoja de Trabajo')
```

```python
print(ot.shape)
ot.head()
```

Se selecciona un subconjunto de variables de interés. Se renombran las columnas para mayor claridad de las variables del conjunto de datos 

```python
ot=ot[['WONUM', 'ELEMENTO', 'WO_NODO', 'WO_FECHACREACION', 'WO_LONG', 'WO_LAT', 'WO_REGION', 'WO_SUBREGION', 'WO_GRUPO_RESPONSABLE', 'WO_CRITICIDAD', 'WO_MUNICIPIO', 'WO_INSTALACION', 'WO_ACTIVIDADPENDIENTE', 'WO_ESTADO']]
ot.rename(columns={'WONUM':'ID', 'WO_ESTADO':'ESTADO', 'WO_FECHACREACION':'FECHA', 'WO_LONG':'COORD_LONG', 'WO_LAT':'COORD_LAT', 'WO_REGION':'REGION', 'WO_SUBREGION':'SUBREGION', 'WO_MUNICIPIO':'MUNICIPIO', 'WO_INSTALACION':'CIRCUITO', 'WO_CRITICIDAD':'CRITICIDAD_NUM', 'WO_NODO':'NODO', 'WO_GRUPO_RESPONSABLE':'GRUPO_RESPONSABLE', 'WO_ACTIVIDADPENDIENTE':'PENDIENTE'}, inplace=True)
print(ot.shape)
```

Se seleccionan solamente las órdenes de trabajo de interes para el análisis, es decir actividades relacionadas con trabajos de poda asociados a las redes de distribución de energía.

```python
ot['PENDIENTE'].unique()
```

Se filtra el conjunto de datos por el campo *PENDIENTE* con el valor *TRABAJOS DE PODA*

```python
ot=ot[ot['PENDIENTE']== 'TRABAJOS DE PODA']
```

```python
print(ot.shape)
ot.head()
```

```python
# Dar formato al campo FECHA
ot['FECHA'] = pd.to_datetime(ot['FECHA'],format='%d/%m/%Y %H:%M:%S')
```

Se crea el campo *Tipo*, con el valor *OT* para identificar el origen del subconjunto de datos (Órdenes de Trabajo)

```python
ot['tipo']='OT'
```

##### **Solicitudes de Servicio**

```python
#Cargar el registro histórico de solicitudes de servicio
#ss = pd.read_excel('/content/drive/MyDrive/ProyectoIntegrador2/exportSS.xlsx', sheet_name='Exportar Hoja de Trabajo')
ss = pd.read_excel('/content/exportSS.xlsx', sheet_name='Exportar Hoja de Trabajo')
```

```python
print(ss.shape)
ss.head()
```

Se seleccionan solamente las solicitudes de servicio de interes para el análisis, es decir actividades relacionadas con trabajos de poda asociados a las redes de distribución de energía.

Se filtra el conjunto de datos por el campo *SS_ACTIVIDADPENDIENTE*

```python
#Eliminar registros sin dato en el campo SS_ACTIVIDADPENDIENTE
ss1=ss.dropna(subset=['SS_ACTIVIDADPENDIENTE'])
```

```python
ss1.shape
```

Se buscan los registros en donde se hace referencia a 'Trabajos de poda' en la actividad pendiente

```python
ss2=ss1[ss1['SS_ACTIVIDADPENDIENTE'].str.contains('Trabajos de poda')]
```

```python
print(ss2.shape)
ss2.head()
```

```python
# Dar formato al campo FECHA
ss2['SS_FECHACREACION'] = pd.to_datetime(ss2['SS_FECHACREACION'],format='%d/%m/%Y %H:%M:%S')
```

###**Crear el conjunto de datos para el análisis**

##### **Crear DataFrame con las Solicitudes de Servicio que tienen asociada una Orden de Trabajo**

Considerando que algunas órdenes de trabajo tienen su origen en una solicitude de servicio, se hace una selección de las **solicitudes de servicio que ya tienen asociada una orden de trabajo** con el fin de no tener eventos repetidos en el conjunto de datos. 

```python
ssot=ss2[ss2['WOHIJA_NUM'].notnull() & ss2['SS_ESTADO'].isin(['ENPROG', 'CERRADA', 'RESUELTA'])]
```

```python
print(ssot.shape)
ssot.head()
```

```python
ssot.columns
```

Se selecciona un subconjunto de variables de interés. Se renombran las columnas para mayor claridad de las variables del conjunto de datos.

```python
ssot=ssot[['WOHIJA_NUM', 'ELEMENTO', 'SS_NODO', 'SS_FECHACREACION', 'SS_LONG', 'SS_LAT', 'SS_REGION', 'SS_SUBREGION', 'SS_GRUPO_RESPONSABLE', 'SS_CRITICIDAD', 'SS_MUNICIPIO', 'SS_INSTALACION', 'SS_ACTIVIDADPENDIENTE', 'WOHIJA_ESTADO']]
```

```python
ssot.rename(columns={'SS_FECHACREACION':'FECHA', 'SS_LONG':'COORD_LONG', 'SS_LAT':'COORD_LAT', 'SS_REGION':'REGION', 'SS_SUBREGION':'SUBREGION', 'SS_MUNICIPIO':'MUNICIPIO', 'SS_INSTALACION':'CIRCUITO', 'SS_CRITICIDAD':'CRITICIDAD_NUM', 'SS_NODO':'NODO', 'SS_GRUPO_RESPONSABLE':'GRUPO_RESPONSABLE', 'SS_ACTIVIDADPENDIENTE':'PENDIENTE', 'WOHIJA_NUM':'ID', 'WOHIJA_ESTADO':'ESTADO'}, inplace=True)
```

Se crea el campo *Tipo*, con el valor *OT_SS* para identificar el origen del subconjunto de datos (Solicitudes de Servicio con Órdenes de Trabajo relacionadas)

```python
ssot['tipo']='OT_SS'
```

```python
ssot.head()
```

##### **Crear DataFrame con las Solicitudes de Servicio que no tienen asociada una Orden de Trabajo**

Se hace una selección de las **solicitudes de servicio que no tienen asociada una orden de trabajo** con el fin de completar el conjunto de datos con la totalidad de los eventos relacionados con trabajos de poda. 

```python
sssinot=ss2[ss2['WOHIJA_NUM'].isnull() & ss2['SS_ESTADO'].isin(['NUEVA', 'RECHAZADA'])]
```

```python
print(sssinot.shape)
sssinot.head()
```

Se selecciona un subconjunto de variables de interés. Se renombran las columnas para mayor claridad de las variables del conjunto de datos.

```python
sssinot=sssinot[['TICKETID', 'ELEMENTO', 'SS_NODO', 'SS_FECHACREACION', 'SS_LONG', 'SS_LAT', 'SS_REGION', 'SS_SUBREGION', 'SS_GRUPO_RESPONSABLE', 'SS_CRITICIDAD', 'SS_MUNICIPIO', 'SS_INSTALACION', 'SS_ACTIVIDADPENDIENTE', 'SS_ESTADO']]
```

```python
sssinot.rename(columns={'SS_FECHACREACION':'FECHA', 'SS_LONG':'COORD_LONG', 'SS_LAT':'COORD_LAT', 'SS_REGION':'REGION', 'SS_SUBREGION':'SUBREGION', 'SS_MUNICIPIO':'MUNICIPIO', 'SS_INSTALACION':'CIRCUITO', 'SS_CRITICIDAD':'CRITICIDAD_NUM', 'SS_NODO':'NODO', 'SS_GRUPO_RESPONSABLE':'GRUPO_RESPONSABLE', 'SS_ACTIVIDADPENDIENTE':'PENDIENTE', 'TICKETID':'ID', 'SS_ESTADO':'ESTADO'}, inplace=True)
```

Se crea el campo *Tipo*, con el valor *SS* para identificar el origen del subconjunto de datos (Solicitudes de Servicio sin Órdenes de Trabajo relacionada)

```python
sssinot['tipo']='SS'
```

```python
sssinot.head()
```

##### **Unir los tres DataFrame en un solo conjunto de datos para el análisis**

```python
df=pd.concat([ot, ssot, sssinot]).reset_index(drop=True)
```

```python
print(df.shape)
df.head()
```

```python
df.dtypes
```

Teniendo en cuenta el contexto de negocio, se establece que la cantidad de eventos correspondientes a trabajos de poda depende en gran medida de la Región geográfica: Clima, tipo de vegetación, geografía, precipitaciones.

```python
df['REGION'].value_counts()
```


Para realizar el análisis se elige inicialmente la Región con una mayor cantidad de eventos, con el fin de disponer de una cantidad mayor de datos.

```python
dfsuro=df[df['REGION']=='SUROESTE']
```

```python
print(dfsuro.shape)
dfsuro.head()
```
